```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import plotly.express as px
import seaborn as sns
```

```python
df = pd.read_csv('Salary_dataset.csv')
```

```python
df.head()
```

```python
df.info()
```

```python
salary = pd.DataFrame(df,columns = ['YearsExperience','Salary'])
```

```python
salary.head()
```

```python
salary.describe()
```

```python
sns.heatmap(salary.corr(), annot=True)
```

```python
fig = px.scatter(salary,x='YearsExperience',y='Salary')
fig.show()
```

```python
x=salary['YearsExperience'].values.reshape(-1,1)
y=salary['Salary'].values.reshape(-1,1)
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y, test_size=0.3)
```

```python
from sklearn.linear_model import LinearRegression
lr = LinearRegression()
```

```python
lr.fit(x_train,y_train)
```

```python
y_pred = lr.predict(x_test)
y_pred
```

```python
y_test
```

```python
plt.scatter(x_test,y_test,c='g',label="Actual Values")
plt.plot(x_test,y_pred,c='r',label="Predicted Values")
plt.xlabel("YearsExperience")
plt.ylabel("Salary")
plt.legend()
plt.show()
```

```python
from sklearn import metrics
print("Mean Absolute Error:",metrics.mean_absolute_error(y_test,y_pred))
print("Mean Squared Error:",metrics.mean_squared_error(y_test,y_pred))
print("Root Mean Squared Error:",metrics.mean_squared_error(y_test,y_pred,squared=False))
print("R2 Score:",metrics.r2_score(y_test,y_pred))
```
